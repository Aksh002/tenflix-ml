# TenFlix V4 — Live Rating-Aware Recommendation Layer

This is a thin interface notebook. Authoritative logic lives in `src/tenflix/v4`. V3 remains available as the benchmark.

```python
from datetime import UTC, datetime
from tenflix.v4.artifacts import load_bundle
from tenflix.v4.config import load_config
from tenflix.v4.events import OnboardingPreferences, RatingEvent, RatingSource
from tenflix.v4.recommender import V4Recommender

config = load_config('configs/v4.yaml')
bundle = load_bundle('artifacts/movielens-v4-eval')
recommender = V4Recommender(bundle)
```

## Preview a new or live user without global retraining

```python
events = [
    RatingEvent(900001, 1, 5.0, datetime.now(UTC), source=RatingSource.ORGANIC),
    RatingEvent(900001, 260, 4.5, datetime.now(UTC), source=RatingSource.ORGANIC),
]
preferences = OnboardingPreferences(preferred_genres=('Adventure', 'Sci-Fi'))
response = recommender.recommend(900001, events, top_k=10, preferences=preferences)
response.recommendations
```
